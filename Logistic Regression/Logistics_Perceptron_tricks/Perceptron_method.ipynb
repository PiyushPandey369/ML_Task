````python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
````

````python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,
                           hypercube=False,class_sep=30)
````

````python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
````

````python
def step(z):
    return 1 if z > 0 else 0 
````

````python
def sigmoid(z):
    return 1/(1+np.exp(-z))
````

````python
def perceptron_tricks(X_train,y_train):
    X_train = np.insert(X_train,0,1,axis=1)
    weights = np.ones(X_train.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,X_train.shape[0])
        y_pred = step(np.dot( X_train[j],weights))
        weights = weights + lr * (y_train[j]-y_pred)*X_train[j]
    return weights[0],weights[1:]
        
````

````python
intercept_, coef_ = perceptron_tricks(X,y)
````

````python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
````

````python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
````

````python
plt.figure(figsize=(8,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
plt.show()
````

````python
model_log = LogisticRegression()
model_log.fit(X,y)
````

````python
m = -(model_log.coef_[0][0]/model_log.coef_[0][1])
b = -(model_log.intercept_/model_log.coef_[0][1])
````

````python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
````

````python
plt.figure(figsize=(8,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
plt.show()
````

````python
def perceptron_sigmoid(X_train,y_train):
    X_train = np.insert(X_train,0,1,axis=1)
    weights = np.ones(X_train.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,X_train.shape[0])
        y_pred = sigmoid(np.dot( X_train[j],weights))
        weights = weights + lr * (y_train[j]-y_pred)*X_train[j]
    return weights[0],weights[1:]
        
````

````python
intercept_1, coef_1 = perceptron_sigmoid(X,y)
m1 = -(coef_1[0]/coef_1[1])
b1 = -(intercept_1/coef_1[1])
x_input11 = np.linspace(-3,3,100)
y_input11 = m1*x_input11 + b1
````

````python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3,label='perceptron')
plt.plot(x_input1,y_input1,color='black',linewidth=3,label='scikit_lib')
plt.plot(x_input11,y_input11,color='green',linewidth=3,label='sigmoid')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.legend()
plt.ylim(-3,2)
plt.show()
````
