```python
import numpy as np 
```

#### Defining input and output features

```python
X = np.array([
    [1,0.5,87.5,100],
    [1,1.5,100,100],
     [1,1.5,75,57.14],
     [1,1,50,80],
     [1,1.5,72.22,75],
     [1,2,77.78,87.50]
])
Y =  np.array([38.84,45.5,35.08,34.62,30.972,39.66])
```

```python
X,Y
```

#### Performing matrix operations and calculating errors

```python
Xt = X.transpose()
```

```python
Xt
```

```python
Xt.shape
```

```python
XtX =  np.dot(Xt,X)
```

```python
XtX.shape
```

```python
XtX
```

```python
XtX_inv = np.linalg.inv(XtX)
XtX_inv
```

```python
final_X = np.dot(XtX_inv,Xt)
final_X
```

```python
 beta = final_X @ Y
```

```python
beta
```

```python
y_pred =  X @ beta
y_pred
```

```python
residuals = Y - y_pred
residuals
```

##### Residual error is the difference between the actual observed value and the value predicted by the regression model.

```python
RSS = np.sum(residuals ** 2)
print(RSS)
```

Mean Squared Error (MSE) is the average of the squares of the residual errors between the actual values and the predicted values in a regression model.

```python
MSE = RSS / 6
print(MSE)
```
