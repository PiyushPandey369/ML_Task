```python
import numpy as np 
import pandas as pd
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('encoded_heart.csv')
```

```python
class My_KNNClassifier:
    
    def __init__(self,neighbour=5):
        self.neighbour = neighbour
        self.X_train = None
        self.y_train = None

    def fit(self,X_train,y_train):
        self.X_train = X_train
        self.y_train = y_train

    def calculate_distance(self,point_A,point_B):
        return np.linalg.norm(point_A-point_B)
        
    def predict(self,X_test):
        y_pred = []
        for i in X_test:
            distance = []
            for j in self.X_train:
                distance.append(self.calculate_distance(i,j))
            n_neighbour = sorted(enumerate(distance),key = lambda x :x[1])[0:self.neighbour]
            label = self.majority_count(n_neighbour)
            y_pred.append(label)
        return np.array(y_pred)

    def majority_count(self,neighbour):
        votes = []
        for i in neighbour:
            votes.append(self.y_train[i[0]])
        votes = Counter(votes)
        return votes.most_common()[0][0]
```

```python
X = df.drop(columns='HeartDisease').values[:]
```

```python
y = df[['HeartDisease']].values[:,0]
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
model = My_KNNClassifier(neighbour=13)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=13)
model_knn.fit(X_train,y_train)
y_pred1 = model_knn.predict(X_test)
accuracy_score(y_test,y_pred1)
```
